```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = np.meshgrid(np.linspace(-2, 2, 100), np.linspace(-2, 2, 100))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

ax1.pcolormesh(x, y, x)
ax2.pcolormesh(x, y, y)

plt.tight_layout()
plt.show()
```

```python
r = np.sqrt(x**2 + y**2)
phi = np.arctan2(y, x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.pcolormesh(x, y, r)
ax2.pcolormesh(x, y, phi)
plt.show()
```

```python
f = np.cos(4*r)**2 + np.sin(4*phi)**2
```

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.pcolormesh(x, y, f)
plt.show()
```

## Part 2

```python
import pooch
url = "https://www.ldeo.columbia.edu/~danielmw/float_data.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
print(files)
```

### 2.1

```python
date = np.load(files[0])
T = np.load(files[1])
S = np.load(files[2])
P = np.load(files[3])
level = np.load(files[4])
lon = np.load(files[5])
lat = np.load(files[6])
```

### 2.2

```python
print(T.shape, S.shape, P.shape)
```

```python
print(lon.shape, level.shape, lat.shape, date.shape)
```

### 2.3

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,5))

ax1.plot(T, level)
ax1.set_xlabel('Temperature')
ax1.invert_yaxis()

ax2.plot(S, level)
ax2.set_xlabel('Salinity')
ax2.invert_yaxis()

ax3.plot(P, level)
ax3.set_xlabel('Pressure')
ax3.invert_yaxis()

plt.tight_layout()
plt.show()
```

### 2.4

```python
T_mean = np.mean(T, axis=1)
T_std = np.std(T, axis=1)

S_mean = np.mean(S, axis=1)
S_std = np.std(S, axis=1)

P_mean = np.mean(P, axis=1)
P_std = np.std(P, axis=1)
```

```python
print(T_mean.shape, S_mean.shape, P_mean.shape)
```

### 2.5

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,5))

ax1.errorbar(T_mean, level, xerr=T_std)
ax1.set_xlabel('Temperature')
ax1.invert_yaxis()

ax2.errorbar(S_mean, level, xerr=S_std)
ax2.set_xlabel('Salinity')
ax2.invert_yaxis()

ax3.errorbar(P_mean, level, xerr=P_std)
ax3.set_xlabel('Pressure')
ax3.invert_yaxis()

plt.tight_layout()
plt.show()
```
